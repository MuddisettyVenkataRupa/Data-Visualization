```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=sns.load_dataset("tips")
```

```python
df.head()
```

```python
df.corr()
```

# Heatmap 

##### Confusion Matrix visualisation using heatmap gives good understanding in Machine Learning Projects

```python
sns.heatmap(df.corr())
```

## JOINT PLOT

```python
sns.jointplot(x='tip',y='total_bill',data=df,kind='hex')
```

```python
sns.jointplot(x = 'tip', y = 'total_bill', data = df, kind = 'reg')
```

## PAIR PLOT

```python
sns.pairplot(df, hue = 'sex')
```

```python
sns.pairplot(df)
```

## DIST PLOT 

```python
sns.distplot(df['tip'], kde = False)
```

```python
## kde - Kernel Gaussian Distribution Estimate
sns.distplot(df['tip'],kde=True,bins=20)
```

# Categorical plots

### COUNT PLOT

```python
sns.countplot('sex',data=df)
plt.show()

## Horizontal bars
sns.countplot(y='sex',data=df)
plt.show()
```

### BAR PLOT

```python
## Bar plot

sns.barplot(y='total_bill',x='sex',data=df)
```

### BOX PLOT 


###### You can observe the outliers in the specified columns using the box plot

```python
sns.boxplot(x = 'smoker', y = 'total_bill', data=df)
```

```python
sns.boxplot(x="day", y="total_bill", data=df,palette='rainbow')
```

```python
sns.boxplot(data = df, orient = 'v')
```

```python
sns.boxplot(x = "total_bill", y ="day", hue = "sex", data = df)
plt.show()
```

### VIOLIN PLOT 


###### Violin plot displays both boxplot and kernel density distribution estimate

```python
sns.violinplot(x = 'total_bill', y = 'day', data = df, palette = 'rainbow')
```
